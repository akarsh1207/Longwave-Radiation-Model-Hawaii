# Importing the Libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
import datetime
```

# Loading the Dataset

```python
df = pd.read_csv('014HI_clean.csv')
```

```python
df
```

# Filtering out columns

```python
print(df.columns)
```

```python
selected_columns = [
    'Date_Time',
    'incoming_radiation_lw_set_1(W/m**2)',
    'solar_radiation_set_1(W/m**2)', 
    'relative_humidity_set_1(%)', 
    'air_temp_set_1(Celsius)', 
    'ELEVATION [ft]', 
]
filter_df = df[selected_columns]
```

```python
filter_df
```

# Processing it to Zoey's | Jiedong's Format 

```python
# Solar Time
filter_df['solartime'] = filter_df['Date_Time'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%dT%H:%M:%SZ').strftime('%Y-%m-%d %H:%M:%S'))
filter_df = filter_df.drop(columns=['Date_Time'])

#Temperature
filter_df['temp'] = filter_df['air_temp_set_1(Celsius)']
filter_df = filter_df.drop(columns=['air_temp_set_1(Celsius)'])

#Altitude/Elevation
filter_df['site_elev'] = filter_df['ELEVATION [ft]'] * 0.3048
filter_df = filter_df.drop(columns=['ELEVATION [ft]'])

#Relative Humidity
filter_df['rh'] = filter_df['relative_humidity_set_1(%)']
filter_df = filter_df.drop(columns=['relative_humidity_set_1(%)'])

#DLW (Downward Longwave)
filter_df['dlw'] = filter_df['incoming_radiation_lw_set_1(W/m**2)']
filter_df = filter_df.drop(columns=['incoming_radiation_lw_set_1(W/m**2)'])

#GHI (Global Horizontal Irradiance)
filter_df['ghi_m'] = filter_df['solar_radiation_set_1(W/m**2)']
filter_df = filter_df.drop(columns=['solar_radiation_set_1(W/m**2)'])
```

```python
filter_df
```

# Supplementary values

```python
df_2022 = pd.read_csv('014HI_2022.csv', skiprows=2)
df_2023 = pd.read_csv('014HI_2023.csv', skiprows=2)
df_2022_2023 = pd.concat([df_2022, df_2023])
```

```python
df_2022_2023['solartime'] = pd.to_datetime(df_2022_2023[['Year', 'Month', 'Day', 'Hour', 'Minute']])
df_2022_2023 = df_2022_2023.drop(columns=['Year', 'Month', 'Day', 'Hour', 'Minute'])
```

```python
df_2022_2023
```

```python
# Convert solartime to datetime in filter_df for accurate merging
filter_df['solartime'] = pd.to_datetime(filter_df['solartime'])

# Merge the dataframes on the 'solartime' column
merged_df = pd.merge(filter_df, df_2022_2023, on='solartime', how='inner')

# Display the merged dataframe
merged_df
```

```python
final_df = merged_df.dropna()
```
